# Step 10

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error, r2_score 
X=df[df.columns[0:4]].values
y=df[df.columns[4:]].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,
                                               test_size=0.15,
                                               shuffle=True)
```

# Step 11

```python
theta = np.matrix(np.array([0,0,0,0]))
alpha = 0.07
iterations = 1000
```

```python
def compute_cost(X, y, theta):
    return np.sum(np.square(np.matmul(X, theta) - y)) / (2 * len(y))
def gradient_descent_multi(X, y, theta, alpha, iterations):
    theta = np.matrix(np.zeros(X.shape[1])).T
    m = len(X)
    gdm_df = pd.DataFrame( columns = ['Bets','cost'])

    for i in range(iterations):
        gradient = (1/m) * np.matmul(X.T, np.matmul(X, theta) - y)
        theta = theta - alpha * gradient
        cost = compute_cost(X, y, theta)
        gdm_df.loc[i] = [theta,cost]

    return gdm_df
```

```python
d=gradient_descent_multi(X_train, y_train, theta, alpha, iterations)
```

```python
gradient_descent_multi(X, y, theta, alpha, iterations).reset_index().plot.line(x='index', y=['cost'])
```

```python
parameters=(d["Bets"].iloc[-1])
```

```python
y_pred_test=np.dot(X_test,parameters)
y_pred_train=np.dot(X_train,parameters)
```

# Step 12

```python
print("MSE of Training set : ",mean_squared_error(y_train,y_pred_train))
print("R2 Score of Training set : ",r2_score(y_train,y_pred_train))
print()
print("MSE of Test set : ",mean_squared_error(y_test,y_pred_test))
print("R2 Score of Test set : ",r2_score(y_test,y_pred_test))
```
